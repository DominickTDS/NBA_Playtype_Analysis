```python
df['Playoff_Team'] = df.Playoff_Team.astype(str)
```

```python
df.head()
```

```python
df['Playoff_Team'] = df['Playoff_Team'].replace(('True','False'),(1,0))
```

```python
df.head()
```

```python
df.Playoff_Team.value_counts()
```

```python
x = df.drop(['Team', 'Season', 'Playoff_Team'], axis = 1)
```

```python
x.head()
```

```python
y = df['Playoff_Team']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

# split the data
X_train, X_val, y_train, y_val = train_test_split(x, y, test_size =0.3, random_state=10)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

from yellowbrick.classifier import ConfusionMatrix
from yellowbrick.classifier import ClassificationReport
from yellowbrick.classifier import ROCAUC
```

```python
model = LogisticRegression()
```

```python
#The ConfusionMatrix visualizer
classes = ['Non Playoffs','Playoffs']
cm = ConfusionMatrix(model, classes=classes, percent=False)
cm.fit(X_train, y_train)
cm.score(X_val, y_val)
for label in cm.ax.texts:
    label.set_size(20)
cm.poof()
```

```python
# Instantiate the visualizer
visualizer = ClassificationReport(model, classes=classes)

visualizer.fit(X_train, y_train)  # Fit the training data to the visualizer
visualizer.score(X_val, y_val)  # Evaluate the model on the test data
g = visualizer.poof()
```

```python
model.score(X_val, y_val)
```
